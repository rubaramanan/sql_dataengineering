### DCL (Data control language)

* Purpose: Control access to the database and database object

Grant: To give user access privileges

Revoke: To takeaway privileges

```python
create user ruba WITH PASSWORD 'hello';

grant select, insert ON all tables in schema public to ruba;
```

```python
psql -Uruba -d demo
```
