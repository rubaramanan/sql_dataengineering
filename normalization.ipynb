Before normalization

After normalization

```python
%%sql

create table student(
    id int primary key,
    name text
);

create table phone(
    student_id int references student(id),
    phone_num text,
    primary key(student_id, phone_num)
);
```

```python
%%sql
insert into student(id, name)
values(1, 'Ruba'),
(2, 'Ramanan');

insert into phone(student_id, phone_num)
values(1, '4001248'),
(2, '6061803'),
(2, '6138702');
```

```python
%%sql
select *
from student as s, phone as p
where p.student_id = s.id;
```
